```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Function to fetch stock data
def get_stock_data(symbol, period='1y'):
    stock = yf.Ticker(symbol)
    history = stock.history(period=period)
    return history

# Function to save stock data to CSV
def save_stock_data_to_csv(symbol, period='1y'):
    stock_data = get_stock_data(symbol, period)

    # Creating a file name based on stock symbol
    filename = f"{symbol}_stock_data.csv"
    stock_data.to_csv(filename)
    print(f"Data for {symbol} saved to {filename}")

# Function to visualize stock data
def plot_stock_data(symbol, period='1y'):
    stock_data = get_stock_data(symbol, period)

    # Plotting the closing price
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Close Price', color='blue')

    plt.title(f'{symbol} Stock Price ({period})')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
if __name__ == '__main__':
    stock_symbol = input("Enter the stock ticker symbol (e.g., AAPL for Apple): ").upper()

    # Fetch and plot stock data
    plot_stock_data(stock_symbol)

    # Save stock data to CSV
    save_stock_data_to_csv(stock_symbol)
```
